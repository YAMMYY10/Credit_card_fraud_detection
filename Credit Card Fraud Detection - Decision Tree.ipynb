## Train Test and Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Decision Tress

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
dtree.fit(X_train,y_train)
```

## Prediction and Evaluation

```python
predictions = dtree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Convert the feature names from pandas Index to a list
feature_names = X.columns.tolist()

plt.figure(figsize=(50, 30))
tree.plot_tree(dtree,
               feature_names=feature_names,  # Use the list of feature names
               class_names=['Class-1', 'Class-0'],
               rounded=True,  # Rounded node edges
               filled=True,  # Adds color according to class
               proportion=True)
plt.show()
```
